```python
import pandas as pd
```

```python
dataset = pd.read_csv("general_data.csv")
```

```python
dataset
```

```python
# checking for null values
dataset.replace(to_replace=['NA',],value=[0],inplace = True)
dataset.isnull()
```

```python
# Dropping rows if any element of row has null value
dataset = dataset.dropna()
dataset
```

```python
dataset.drop_duplicates()
```

> Now we go with our main objective to compare attrition with remaining fields

```python
dataset.columns
```

```python
print("Attrition:",dataset.Attrition.unique())
print("BusinessTravel:",dataset.BusinessTravel.unique())
print("Department:",dataset.Department.unique())
print("EducationField:",dataset.EducationField.unique())
print("Gender:",dataset.Gender.unique())
print("JobRole:",dataset.JobRole.unique())
print("MaritalStatus:",dataset.MaritalStatus.unique())
print("Over18:",dataset.Over18.unique())
```

```python
dataset.replace(to_replace=['Yes','No'],value=[1,0],inplace = True)
dataset.replace(to_replace=['Travel_Rarely', 'Travel_Frequently', 'Non-Travel'],value=[2,1,0],inplace = True)
dataset.replace(to_replace=['Sales', 'Research & Development', 'Human Resources'],value=[2,1,0],inplace = True)
dataset.replace(to_replace=['Life Sciences', 'Other', 'Medical', 'Marketing', 'Technical Degree',
 'Human Resources'],value=[5,4,3,2,1,0],inplace = True)
dataset.replace(to_replace=['Female', 'Male'],value=[1,0],inplace = True)
dataset.replace(to_replace=['Healthcare Representative', 'Research Scientist', 'Sales Executive',
 'Human Resources', 'Research Director', 'Laboratory Technician',
 'Manufacturing Director', 'Sales Representative', 'Manager'],value=[8,7,6,5,4,3,2,1,0],inplace = True)
dataset.replace(to_replace=['Married', 'Single', 'Divorced'],value=[2,1,0],inplace = True)
dataset.replace(to_replace=['Y', 'N'],value=[1,0],inplace = True)
```

```python
dataset
```

```python
from scipy.stats import pearsonr
```

```python
dataset
```

```python
dataset.corr()
```

```python
dataset.columns
```

```python
stat1,p1=pearsonr(dataset.Attrition,dataset.Age)
print(stat1,p1)
```

```python
stat2,p2=pearsonr(dataset.Attrition,dataset.DistanceFromHome)
print(stat2,p2)
```

```python
for i in dataset.columns :
    if i not in ['Attrition','EmployeeID','EmployeeCount','StandardHours','NumCompaniesWorked','TotalWorkingYears'] :
        temp,p = pearsonr(dataset.Attrition,dataset[i])
        print(i,end=": ")
        print(temp,p,end="\n\n")
```

- There is no field with perfect positive or negative correlation
- Since p of Age is < 0.05 H0 is rejected and can conclude they are correlated
- Since p of BusinessTravel is > 0.05 H0 is accepted i.e; both are correlated
- Since p of Department is < 0.05 H0 is rejected
- ........ and remaining can be concluded as above
